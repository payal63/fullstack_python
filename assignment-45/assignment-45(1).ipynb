```python
# Calculate the average house price using .mean()
avg_price=house_price['price'].mean()
avg_price=round(avg_price,2)
avg_price
print(f" Average price is : {avg_price}")
```

```python
#Find the most common number of bedrooms 
common_bedroom=house_price['bedrooms'].mode()
print(f" common bedroom is : {common_bedroom}")
```

```python
# find the most common bathrooms
common_bathroom= house_price['bathrooms'].mode()
print(f" common bathroom is : {common_bathroom}")
```

```python
#Identify any correlation between the house price and other numeric features (e.g., sqft_living, waterfront, and view) using .corr() functions
correlation= house_price['price'].corr(house_price['sqft_living'])
correlation= round(correlation,2)
correlation
print(f" correlation is {correlation} , It is moderate positive relation")
```

```python
#correlation between price and waterfront
correlation=house_price['price'].corr(house_price['waterfront'])
correlation=round(correlation,2)
correlation
print(f" correlation is {correlation}, it is weak positive correlation")
```

```python
#correlation betweem price and view
correlation=house_price['price'].corr(house_price['view'])
correlation=round(correlation,2)
correlation
print(f" correlation is {correlation}: It is weak positive correlation")
```

```python
#Calculate the average price of non-waterfront houses
average_price = house_price[house_price['waterfront'] == 0]['price'].mean()
average_price=round(average_price,2)
print(f"Average price of non-waterfront houses is : {average_price}")
```

```python
# calculate the average price of waterfront houses
average_price=house_price[house_price['waterfront']==1]['price'].mean()
average_price=round(average_price,2)
print(f"average price of waterfront houses is : {average_price}")
```

```python
#Determine the average price of houses with different numbers of floor
avg_price_by_floors = house_price.groupby('floors')['price'].mean()
avg_price_by_floors = round(avg_price_by_floors,2)
print(f" Average price of houses with diff. number of floor is : {avg_price_by_floors}")
```

```python
#Analyze the average price based on the house condition
avg_price_by_condition=house_price.groupby('condition')['price'].mean()
avg_price_by_condition =round(avg_price_by_condition,2)
print(f"Average price on condition is : {avg_price_by_condition}")
```
